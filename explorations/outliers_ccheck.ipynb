```python
import pandas as pd

df = pd.read_csv("../src/taxipred/data/cleaned_data.csv")

df.head()
```

```python
df.describe().T
```

```python
import numpy as np


# Beräkna 91:e och 1:a percentilerna
upper_limit = df['Trip_Price'].quantile(0.99)
lower_limit = df['Trip_Price'].quantile(0.01)

# Skapa en ny DataFrame utan outliers 
df_filtered = df[(df['Trip_Price'] <= upper_limit) & (df['Trip_Price'] >= lower_limit)]

print(f"Antal rader i originaldata: {len(df)}")
print(f"Antal rader efter borttagning av outliers: {len(df_filtered)}")
```

```python
# Exporterar DataFrame till en ny CSV-fil

# df_filtered.to_csv('cleaned_data_no_outliers.csv', index=False)

# Exportera en till fil med weather för att kunna visa korrelation mellan väder och trip duration
df_filtered.to_csv('cleaned_data_no_outliers.csv', index=False)
```
